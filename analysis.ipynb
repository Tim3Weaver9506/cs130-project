```python
def bob_ross_episode_list():
    new_list = []
    for line in open("elements-by-episode(1).csv"):
        clean_line = line.strip()
        clean_list = clean_line.split(',')
        new_list.append(clean_list)
    return new_list
bob_ross_episode_list()
```

```python
episode_data = bob_ross_episode_list()
episode_data[1]
```

```python
import pandas as pd

bobross_pd = pd.read_csv("elements-by-episode(1).csv")
```

```python
bob_ross_columns = bobross_pd.columns[2:]
```

```python
def episode_element():
    episode_dict = {}
    for line in episode_data[1:]:
        for column in bob_ross_columns:
            if line[bobross_pd.columns.get_loc(column)] == '1':
                episode_dict.setdefault(line[0], []).append(column)
    return episode_dict
episode_element()
```
