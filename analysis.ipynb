```python
def bob_ross_episode_list():
    new_list = []
    for line in open("elements-by-episode(1).csv"):
        clean_line = line.strip()
        clean_list = clean_line.split(',')
        new_list.append(clean_list)
    return new_list
```

```python
episode_data = bob_ross_episode_list()
episode_data[1]
```

```python
import pandas as pd

bobross_pd = pd.read_csv("elements-by-episode(1).csv")
bobrosspaintings_pd = pd.read_csv("bob_ross_paintings.csv")
```

```python
bob_ross_columns = bobross_pd.columns[2:]
color_columns = bobrosspaintings_pd.columns[11:]
```

```python
episode_dict = {}
for line in episode_data[1:]:
    for column in bob_ross_columns:
        if line[bobross_pd.columns.get_loc(column)] == '1': # Check if element is present
            episode_dict.setdefault(line[0], []).append(column)
episode_dict

```

```python
def episode_element_finder():
    question = input("Enter episode season + episode number (e.g., S01E01): ")
    episode_dict.get(question) 
    if question in episode_dict:
        print(f"Elements in {question}: {', '.join(episode_dict[question])}")
    else:
        print("Episode not found, please try again.")
```

```python
episode_element_finder() #Using Season 1 Episode 1
```

```python
def element_in_episodes():
    element_question = input("Enter an element to find which episodes it appears in: ")
    episodes_with_element = []
    for episode, elements in episode_dict.items():
        if element_question.upper() in elements:
            episodes_with_element.append(episode)
    if episodes_with_element:
        print(f"The element '{element_question.upper()}' appears in the following episodes: {', '.join(episodes_with_element)}")
    else:
        print(f"The element '{element_question.upper()}' does not appear in any episodes.")
```

```python
element_in_episodes()
```
