# Lists

Like a string, a **list** is a sequence of values.

| Data type | Sequence of values | Surrounded by... |
------------|--------------------|------------------|
| String | Characters | `'` or `"` |
| List   | Any type, called elements | `[` and `]` |

A list of integers:

```python
numbers = [42, 123]
```

A list of strings:

```python
cheeses = ['Cheddar', 'Swiss', 'Gouda']
```

The elements of a list don't have to be the same type.

The following list contains a string, a float, an integer, and even another **nested** list.

```python
t = ['spam', 2.0, 5, [10, 20]]
```

A list that contains no elements is an empty list.

```python
empty = []
```

The `len` function returns the length of a list.

```python
len(cheeses)
```

```python
len(empty)
```

Nested lists count as a single element.

```python
t = ['spam', 2.0, 5, [10, 20]]
len(t)
```

# Mutability

Unlike strings, lists are mutable (changeable).

Otherwise, there are many similarities:
* Zero indexed
* Put index in brackets to access elements
* Negative indexes work backwards

```python
word = 'Parmesan'
word[0] = 'F'
```

```python
cheeses
```

```python
cheeses[0] = 'Pepperjack'
cheeses
```

The `in` operator also works on lists. It checks whether a given element appears anywhere in the list.

```python
'Swiss' in cheeses
```

```python
'Provolone' in cheeses
```

`10` is not an element of `t` because it is an element of a nested list.

```python
t = ['spam', 2.0, 5, [10, 20]]
10 in t
```

# List slices

List slices are the same as string slices.

```python
letters = ['a', 'b', 'c', 'd']
letters[1:3]
```

```python
letters[:2]
```

```python
letters[2:]
```
