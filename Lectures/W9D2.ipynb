# Last assignment
Write a function named has_abc that checks to see if the provided parameter is a string. If it is not, return None. If it is a string, see if it contains the letters a, b, or c (checks for uppercase letters are not required). If it does, return True. Otherwise, return False.

```python
def has_abc(word):
    if not isinstance(word, str):
        return None
    else:
        for letter in word:
            if letter == 'a' or letter == 'b' or letter == 'c':
                return True
        return False
```

```python
has_abc(42)
```

```python
print(has_abc(42))
```

```python
has_abc("42")
```

```python
has_abc("abc")
```

```python
has_abc("dog")
```

```python
has_abc("happy")
```

```python
has_abc("xyzac")
```

# Updating variables

You're allowed to make more than one assignment to the same variable.

```python
x = 5
print(x)
```

```python
x = 7
print(x)
```

x no longer refers to 5, it refers to 7 now.

This is called **updating** the variable.

```python
x = 7
print('before update:',x)

x = x + 1
print('after update:',x)
```

`x = x + 1` means:
* get the current value of `x`
* add one
* assign the result back to `x`

Why doesn't the code below work?

```python
z = z + 1
```

We need to **initialize** the variable before we can use it.

```python
z = 0
z = z + 1
print(z)
```

The practice of increasing the variable is called **incrementing**.

The practice of decreasing the variable is called **decrementing**.

These are so common, that we have shortcuts (`-=` and `+=`).

```python
z = 0
z = z + 1
print(z)

z = 0
z += 1
print(z)
```

Write a program that asks the user for 5 numbers and counts how many of them are even. To do this, follow these steps:
* Initialize a counter variable named `even_count`
* Use a loop to ask for input 5 times
* Increment the counter each time the user enters an even number

This exercise will use the following concepts from prior classes:

###### Solution

```python
even_count = 0

for ask in range(5):
    number = int(input('Enter a number:'))
    if number % 2 == 0:
        even_count += 1

print(even_count)
```
