# Important function follow up

```python
def rectangle_area(width, height):
    area = width * height
    print(area)
```

```python
rectangle_area(5, 10)
rectangle_area(7, 3)
rectangle_area(3, 6)
```

Variable names in function calls have nothing to do with the name in the function itself:

```python
apple = 5
banana = 10
rectangle_area(apple, banana)
```

Functions can call other functions, which complicates the flow of execution:

```python
def rectangle_area(width, height):
    area = width * height
    print(area)

def compare_rectangles(w1, h1, w2, h2):
    print("First rectangle:")
    rectangle_area(w1, h1)               # Calls the other function
    print("Second rectangle:")
    rectangle_area(w2, h2)               # Calls the other function
```

```python
compare_rectangles(2,3,6,2)
```

# Functions calling other functions & looping

The song "99 Bottles of Beer" starts with this verse:

> 99 bottles of beer on the wall  
> 99 bottles of beer  
> Take one down, pass it around  
> 98 bottles of beer on the wall  

Then the second verse is the same, except that it starts with 98 bottles and ends with 97. The song continues for a very long time until there are 0 bottles of beer.

Write a function called `bottle_verse` that takes a number as a parameter and displays the verse that starts with the given number of bottles.

```python
def bottle_line(n, suffix):
    print(n, 'bottles of beer', suffix)

def bottle_verse(n):
    bottle_line(n, 'on the wall')
    bottle_line(n, '')
    print('Take one down, pass it around')
    bottle_line(n-1, 'on the wall')
```

```python
bottle_verse(5)
```

This `for` loop counts down from `99` to `1` and prints the whole song.

But don't worry about fully understanding it now.

```python
for n in range(99, 0, -1):
    bottle_verse(n)
```

## Repetition

Here's a simpler `for loop`.

```python
for i in range(3):
    print(i)
```

* Indented just like a function
* Starts with keyword of `for` instead of `def`
* Colon just like a function
* What is `i`, `in`, and `range()`?

`range(3)` is essentially the same as a list `[0, 1, 2]`

This example helps us understand what `i` and `in` do as well...

```python
for i in [0, 1, 2]:
    print(i)
```

The code "loops" 3 times:
1. First loop: `i` is `0`
2. Second loop: `i` is `1`
3. Third loop: `i` is `2`

Why do we start at `0`? Or more importantly... why does `range(3)` start at `0`?
