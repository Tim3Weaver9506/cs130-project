# List review

In a list, the indices have to be integers. 

```python
animals = ['alligator', 'bat', 'camel']
```

```python
animals[1]
```

We can add to the list using `append`.

```python
animals.append('dolphin')
animals
```

# Dictionaries

The curly braces, `{}`, represent an empty dictionary.

```python
animals = {}
animals
```

Instead of using `append` like we did with lists, we do the following.

Confusingly, we use square brackets to add items to the dictionary.

```python
animals['a'] = 'alligator'
```

But why do I see both `'a'` and `'alligator'` in that line of code?

A dictionary is like a list, but more general in that it doesn't have to use integer indexes.

The above assignment adds an item to the dictionary, which represents the association of a **key** and a **value**.

In this example, the key is the string `'a'` and the value is the string `'alligator'`. The key and value separated by a colon:

```python
animals
```

We can add more items like this.

```python
animals['b'] = 'bat'
animals['c'] = 'camel'
animals
```

Now the dictionary contains three items.

To look up a key and get the corresponding value, we use the bracket operator.

```python
animals['b']
```

If the key isn't in the dictionary, we get a `KeyError`.

```python
animals['f']
```

The `len` function works on dictionaries; it returns the number of items.

```python
len(animals)
```

You can also create dictionaries like this, using `{` and `}`.

```python
animals = {'a': 'alligator', 'b': 'bat', 'c': 'camel'}
```

Each item consists of a key and a value separated by a colon.

# The in operator

The `in` operator works on dictionaries, too; it tells you whether something appears as a ***key*** in the dictionary.

```python
'b' in animals
```

The `in` operator does *not* check whether something appears as a value.

```python
'bat' in animals
```

To see whether something appears as a value in a dictionary, you can use the method `values`, which returns a sequence of values, and then use the `in` operator.

```python
'bat' in animals.values()
```

# A collection of counters

Suppose you are given a string and you want to count how many times each letter appears.

```python
counter = {}
```

As we loop through the letters in the string, if we find the letter `'a'` we'd add it to the dictionary:

```python
counter['a'] = 1
counter
```

Increment the counter for a every time we see it again.

```python
counter['a'] += 1
counter
```

The following function uses these features to count the number of times each letter appears in a string.

```python
def value_counts(string):
    counter = {}
    for letter in string:
        if letter not in counter:
            counter[letter] = 1
        else:
            counter[letter] += 1
    return counter
```

Each time through the loop, if `letter` is not in the dictionary, we create a new item with key `letter` and value `1`.
If `letter` is already in the dictionary we increment the value associated with `letter`.

```python
counter = value_counts('brontosaurus')
counter
```

The items in `counter` show that the letter `'b'` appears once, `'r'` appears twice, and so on.
