# Integer division and modulus

Recall that the integer division operator, `//`, divides two numbers and rounds down to an integer.

```python
minutes = 105 # movie run time
minutes / 60  # how many hours?
```

But, we don't write hours with decimal points.

```python
minutes = 105
hours = minutes // 60 # rounds down
remainder = minutes - (hours * 60)
print(hours, "hour")
print(remainder, "minutes")
```

Or you could use the **modulus operator**, `%`, which divides two numbers and returns the remainder.

```python
remainder = minutes % 60
print(remainder)
```

The modulus operator can also check whether one number is divisible by another: if `x % y` is zero, then `x` is divisible by `y`.

```python
123 % 10
```

```python
120 % 10
```

# Boolean expressions

A **boolean expression** is an expression that is either true or false.

The `==` operator compares two values. It produces `True` if they are equal and `False` otherwise.

```python
5 == 5
```

```python
5 == 7
```

Remember that `=` assigns a value to a variable and `==` compares two values. 

```python
x = 5
y = 7
```

```python
x == y
```

`True` and `False` are special values that belong to the type `bool`;
they are not strings:

```python
type(True)
```

```python
type(False)
```

```python
type("True")
```

Remember the modulus operator can  check whether one number is divisible by another: if `x % y` is zero, then `x` is divisible by `y`.

```python
123 % 10 == 0
```

```python
120 % 10 == 0
```

The `==` operator is one of the **relational operators**; the others are:

```python
x != y               # x is not equal to y
```

```python
x > y                # x is greater than y
```

```python
x < y               # x is less than to y
```

```python
x >= y               # x is greater than or equal to y
```

```python
x <= y               # x is less than or equal to y
```
