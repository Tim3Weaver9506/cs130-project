Write a program that asks the user for 5 numbers and counts how many of them are even. To do this, follow these steps:
* Initialize a counter variable named `even_count`
* Use a loop to ask for input 5 times
* Increment the counter each time the user enters an even number
* Print the total number of even numbers at the end

# The in operator

Our previously created `has_abc()` function was more complicated than it needed to be, because we have an operator `in` that checks whether a character appears in a string.

```python
def has_abc_original(word):
    if not isinstance(word, str):
        return None
    else:
        for letter in word:
            if letter == 'a' or letter == 'b' or letter == 'c':
                return True
        return False
```

```python
def has_abc_in(word):
    if not isinstance(word, str):
        return None
    else:
        if 'a' in word or 'b' in word or 'c' in word:
            return True
        else:
            return False
```

The above function uses a nested conditional, which isn't ideal. We're also redundantly creating booleans. We could rewrite it as follows:

```python
def has_abc_in(word):
    if not isinstance(word, str):
        return None
    else:
        return 'a' in word or 'b' in word or 'c' in word
```

```python
has_abc_in("HaLLoWeeN")
```

```python
has_abc_in("HALLoWeeN")
```

Let's improve our code to also handle uppercase. We can use the `lower()` function.

```python
word = "HaLLoWeeN"
print(word)
print(word.lower())
print(word)
```

```python
def has_abc_in_lower(word):
    if not isinstance(word, str):
        return None
    else:
        word = word.lower()
        return 'a' in word or 'b' in word or 'c' in word
```

```python
has_abc_in_lower("HALLoWeeN")
```

# Search

Let's write a more general use function `uses_any` that takes a second parameter that is a string of letters to search for. We'll leave off `isinstance()` to simplify for now.

```python
def uses_any(word, letters):
    for letter in word.lower():
        if letter in letters.lower():
            return True
    return False
```
