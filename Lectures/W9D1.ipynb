# Checking types

One way to improve the reliability of our programs is to make sure a variable is the data type we expect.

The built-in function `isinstance` can help us with that.

```python
isinstance(3, int)
```

```python
isinstance(3.1, int)
```

```python
def bigger(num_one, num_two):
    if not isinstance(num_one, int) or not isinstance(num_two, int):
        print("Must provide integers")
        return None
    elif num_one >= num_two:
        return num_one
    else:
        return num_two
```

```python
bigger(4, 7)
```

```python
returned = bigger(3.3, 8)
print(returned)
```

Notice the ordering of the print statement before the return statement above.

# For loops

We've already used these with the `range` function as follows:

```python
for i in range(3):
    print(i)
```

We can also use `for loops` to iterate through strings one letter at a time.

```python
for letter in "hello":
    print(letter)
```

Let's be a little more fancy.

```python
for letter in "hello":
    if letter == 'e':
        print("Got an E")
```

We'll put it in a function now. Notice the indentation of `return False`.

```python
def has_e(word):
    for letter in word:
        if letter == 'e':
            return True
    return False
```

```python
has_e("lane")
```

```python
has_e("pill")
```
