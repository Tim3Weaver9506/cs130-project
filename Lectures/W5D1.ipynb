# Defining & calling functions

We can write (**define**) our own functions! This function takes no arguments, and prints two strings.
* How do we know it takes no arguments?
* Why is nothing printed to the screen when we run it?

```python
def print_lyrics():
    print("I'm a lumberjack, and I'm okay.")
    print("I sleep all night and I work all day.")
```

Important syntax points:
* `def` keyword
* Colon after any arguments in `()`
* Must indent (4 spaces/tab)

We call it like any other function.

```python
print_lyrics()
```

# Functions that accept parameters (arguments)

We can send two arguments to the `math.pow()` function.

```python
math.pow(5, 2)
```

Our own functions can accept arguments as well. They're called **parameters** when we're defining the function.

```python
def print_twice(phrase):
    print(phrase, phrase)
```

```python
print_twice("Yay for functions!")
```

Let's **generalize** that function a bit (to make it less specific than printing twice).

```python
def print_times(phrase, times):
    print((phrase + " ") * times)
```

```python
print_times("Yay for functions!", 6)
```

# A function's usefulness

But, why? Why not just write the code below instead? It outputs the same thing.

```python
print("Yay for functions!", "Yay for functions!")
```

This is a super important question, and is a core computer science/programming concept! Let's see why with another example.

```python
# Calculate the area of a rectangle
width1 = 5
height1 = 10
area1 = width1 * height1
print(area1)

# Calculate the area of another rectangle
width2 = 7
height2 = 3
area2 = width2 * height2
print(area2)

# Calculate the area of a third rectangle
width3 = 3
height3 = 6
area3 = width3 * height3
print(area3)
```

The above code involves a lot of copying & pasting and is redundant. That is a clue that a function would be useful.

```python
def rectangle_area(width, height):
    area = width * height
    print(area)

rectangle_area(5, 10)
rectangle_area(7, 3)
rectangle_area(3, 6)
```
