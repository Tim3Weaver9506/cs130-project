# Repetition in functions

Last class we discussed a function like `print_twice`

```python
def print_twice(phrase):
    print(phrase)
    print(phrase)

print_twice("Yay for functions!")
```

We **generalized** it by making it print more than two times specifically. We can now do that with a for loop instead of string concatenation (like we did last class).

```python
def print_times(phrase, times):
    for i in range(times):
        print(phrase)

print_times("Yay for for loops!", 6)
```

Note above that we're not required to use the `i`, and we can name it whatever we want.

# Variables and parameters are local

* When you create a variable inside a function, it is **local** to that function
* It only exists inside the function
* It cannot be accessed outside the function

```python
def cat_print(part1, part2):
    cat = part1 + part2
    print(cat)

line1 = 'Always look on the '
line2 = 'bright side of life.'
cat_print(line1, line2)

print(cat)
```

The local variable named `cat` is destroyed when the function ends.

Parameters are also local. `part1` or `part2` do not exist outside of the function either.

```python
print(part1)
```
