```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #This is the API to matplotlib
housing = pd.read_csv("sample_data/california_housing_test.csv")
```

```python
#median_income is currently measured in tens of thousands dollars
housing.median_income
```

```python
#I find that confusing, so lets see it in dollars
def time_ten_thousand(d):
    return d * 10000

housing.median_income.apply(time_ten_thousand)
```

# Maps
A map is the [name](https://en.wikipedia.org/wiki/Map_(higher-order_function)) of a higher-order function that applies a given function to each element of a collection, e.g. a list or set, returning the results in a collection of the same type. It is often called *apply-to-all* when considered in functional form.

We'll use a **lambda function**, which is an anonymous function.

```python
#Instead of writing a single use function, we can instead use a lambda function
housing.median_income.map(lambda d: d * 10000)
```

```python
#Important note: map() does *NOT* change the original data frame
housing.head()
```

```python
#You could always create a new column if you want
housing['median_income_ones'] = housing.median_income.map(lambda d: d * 10000)
housing.head()
```

# Matplotlib
* Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
* Here is a simple example from our data, which we'll explore in more detail below:

```python
housing.median_income.plot(kind='hist')
plt.title('Frequency of Median Income')
plt.xlabel('Dollars')
plt.show()
```

```python
housing.plot(kind='scatter', x='population', y='median_income')
plt.title('Population by Median Income')
plt.xlabel('Population')
plt.ylabel('Median Income ($)')
plt.show()
```

```python
#Improve readability by making it bigger and the dots smaller
housing.plot(kind='scatter', x='population', y='median_income', figsize=(12,6), s=1)
plt.title('Population by Median Income')
plt.xlabel('Population')
plt.ylabel('Median Income ($)')
plt.show()
```
