# String comparison

```python
'orange' == 'orange'
```

```python
# a is before o
'apple' < 'orange'
```

```python
# p is not before o
'pineapple' < 'orange'
```

```python
# uppercase letters come before lowercase
'Pineapple' < 'orange'
```

```python
# always converting is best
'Pineapple'.lower() < 'orange'.lower()
```

```python
'Pineapple'.upper() < 'orange'.upper()
```

`lower()` and `upper()` are not functions, they're **methods** that are only available to use with strings. Therefore we don't call them like this:

```python
upper(demo_word)
```

Instead we use **dot operator** to call them like this:

```python
demo_word = 'Pineapple'
demo_word.upper()
```

In this case, the method takes no arguments. This method only works with strings...

```python
demo_int = 45
demo_int.upper()
```

# Other string methods

startswith:

```python
for line in open('words.txt'):
    if line.startswith('zebr'):
        print(line.strip())
```

endswith:

```python
for line in open('words.txt'):
    line = line.strip()
    if line.endswith('bre'):
        print(line)
```

count (this method does take an argument):

```python
word = 'associate'
word.count('s')
```

```python
for line in open('words.txt'):
    line = line.strip()
    if line.count('a') > 4:
        print(line)
```

# Breaking a loop

This code will print more words than we'd like to the screen making us scroll too much. We'll fix this problem by rewriting it below to only display a couple of words using a counter and a `break` statement.

```python
for line in open('words.txt'):
    if line.startswith('z'):
        print(line.strip())
```

```python
line_num = 0
for line in open('words.txt'):
    if line.startswith('z'):
        print(line.strip())
        line_num += 1
    if line_num == 5:
        break
```
