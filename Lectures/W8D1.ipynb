# Some functions return values

```python
number = 3.14
round(number)
```

We can assign what the function returns to a variable.

```python
number = 3.14
rounded_number = round(number)
print(rounded_number)
```

But the functions we've written so far don't return anything.

```python
def print_times(phrase, times):
    print(phrase * times)

print_times("Hello there! ", 4)
```

```python
repeated_string = print_times("Hello there! ", 4)
print(repeated_string)
```

But we can also return a value!

```python
def return_times(phrase, times):
    return(phrase * times)

repeated_string = return_times("Hello there! ", 4)
print(repeated_string)
```

Since it's now in a variable, we can use it again later!

```python
print("Repeated String:",repeated_string)
```

A function like this is called a **pure function** because it doesn't display anything or have any other effect -- other than returning a value.

## And some functions return None

If a function doesn't have a `return` statement, it returns `None`, which is a special value like `True` and `False`.

This function uses the `print` function to display a string, but it does not use a `return` statement to return a value.

If we assign the result to a variable, it displays the string anyway. 

```python
result = print_times("Hello there! ", 4)
```

And if we print the value, we get `None`:

```python
print(result)
```

## Return values and conditionals

We can use conditionals to apply logic to what is returned.

```python
def num_longer(str_one, str_two):
    if len(str_one) == len(str_two):
        return "Equal"
    elif len(str_one) > len(str_two):
        return len(str_one) - len(str_two)
    else:
        return "Error"
```

```python
example1 = num_longer("Blue","Ship")
example2 = num_longer("Blue","Hat")
example3 = num_longer("Blue","Banana")
print("Blue/Ship:",example1)
print("Blue/Hat:",example2)
print("Blue/Banana:",example3)
```
