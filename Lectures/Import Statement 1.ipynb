# The import statement
* Not all functionality is available to us by default in Python.
* We can add additional features by importing code you (or other people) wrote.
* For example, the following statement imports the `math` package (also referred to as modules or libraries).

````python
import math
````

* A package is a collection of variables and functions.
* They become available to us after we import the package (like checking a book out of the library and suddenly having immediate access to all the knowledge within it).
* Here are some examples:

````python
math.pi
````

````python
math.sqrt(25)
````

````python
import random
random.randint(1, 10)
````

````python
import time
time.sleep(5)
print("Done!")
````

# Expressions and statements

So far, we've seen a few kinds of **expressions**. An expression can be a single value, like an integer, floating-point number, or string.

````python
"Hello"
````

An **expression** can also be a collection of values and operators. And it can include variable names and function calls.

````python
19 + n + round(math.pi) * 2
````

A **statement** is a unit of code that has an effect, but no value like this:

````python
n = 17
````

Similarly, an import **statement** has an effect but it is not visible.

````python
import math
````

| | Expression | Statement |
|-|------------|-----------|
|Produces a value?| Yes | No |
|An instruction?|No|Yes|
|Description|Can be part of a statement|A command|
|Example|2 + 3|y = 2 + 3|
|Print directly?|Yes|Sometimes|
|Term for when it's run|Evaluation|Execution|
